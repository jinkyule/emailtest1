```python
import streamlit as st

st.title("🔒 Site B - 이메일 게시판 (AI 난독화)")

st.write("이메일 주소는 사람에게는 보이지만 크롤러에겐 어렵게 표시됩니다.")

emails = [
    {"name": "test", "domain": "example", "tld": "com"},
    {"name": "hello", "domain": "company", "tld": "com"},
    {"name": "user123", "domain": "domain", "tld": "net"},
]

for email in emails:
    display = f"{email['name']}@{email['domain']}.{email['tld']}"
    html_email = f"<span>{email['name']}<span style='display:none'>REMOVED</span>@{email['domain']}<span style='display:none'>HIDDEN</span>.{email['tld']}</span>"
    st.markdown(html_email, unsafe_allow_html=True)
```
